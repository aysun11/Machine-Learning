```python
veri={"isim":["ali","can","buse","alp","nur"],
      "puan":[90,80,70,87,65],
      "spor":["futbol","kayak","yüzme","tenis","karate"],
      "cinsiyet":["E","E","K","E","K"]}
df=pd.DataFrame(veri)
df
```

```python
df.tail() # son 5 satırı görmek içn
```

```python
# sütunların sırası değiştirme  ve  indexlere etiket atama
df=pd.DataFrame(veri,columns=["isim","spor","cinsiyet","puan"],index=["bir","iki","uc","dort","bes"])
df
```

```python
#  yaas sütunu eklendi
df=pd.DataFrame(veri,columns=["isim","spor","cinsiyet","puan","yas"])
df
```

```python
# belirli bir sütunu almak için
sutunlar=["isim","spor"]
df[sutunlar]
```

```python
# istediğin satırı çağırma
df.loc[["bir","iki"]]
```

```python
# yaş değişkenine değer atma
df["yas"]=18
# yada
deger=[17,15,16,27,22]
df["yas"]=deger
df
```

```python
# sütun oluşturup deger atama
df["gecti"]=df.puan>70
df
```

indexing - selection - filtering

```python
import numpy as np
```

```python
obje=pd.Series(np.arange(5),index=["a","b","c","d","e"])
obje[0:3]
```

```python
obje[["a","c"]]  # sadece a ve c nin değerlerni getir
# ya da
obje[[0,2]]
# ya da
obje["a":"c"]
# ya da
obje["a":"c"]
```

```python
# indexleme
verii=pd.DataFrame(np.arange(16).reshape(4,4),index=["bursa","ankara","rize","malatya"],
                  columns=["bir","iki","uc","dort"])
verii
```

```python
# iloc için numaralar kullanılır
# loc için etiketler kullanılır
# birinci indeksin satır ve sütunlarını görmek için
verii.iloc[1]
# belirli birinci indeksin satırın belirli sütünlarını seçmek için
verii.iloc[1,[1,2,3]]
```

```python
# bir satırın birden fazla sütunun seçmek
verii.loc["rize",["bir","iki"]]
```

```python
# reindex metodu indexleri yeniden düzenler
s=pd.Series([1,2,3,4],index=["a","b","c","d"])
s2=s.reindex(["b","d","a","c","e"])
s2
# boş değerler atama yapbilir
s3=pd.Series(["mavi","sari","mor"],index=[0,2,4])
s3.reindex(range(6),method="ffill")
```

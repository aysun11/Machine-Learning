```python
import matplotlib.pyplot as plt
%matplotlib inline
 # görselleştirmg yapmak için yazılır
```

```python
x=range(1,8)
y=[50,51,52,48,47,49,46]
```

```python
plt.plot(x,y)  #çizdirme
```

```python
plt.suptitle("en üst başlık")
plt.xlabel("Haftanın Günleri")
plt.ylabel("Sıcaklık Değerleri")
plt.title("Sıcaklığın Günlere Dağılımı")

# grafiğin rengini değiştirme color sayıda verilebilir #203040 gibi
# çizgi kalınlığını değiştirme  linewidth
# çizginin nasıl olacağını belirler  l,inestyle
# çizginin şefaflığını belirleme
plt.plot(x,y,color="blue",linewidth=3, linestyle=":",alpha=0.4)
# ya da aşağıdaki gibi yazılır rengin ilk harfi kalınlığı ve şekili
#plt.plot(x,y, "k3-.")
```

```python
days=range(1,8)
max_t=[50,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
```

```python
plt.plot(days,max_t)
```

```python
# üç tane veriyi aynı gfakte çizdirme
# hangi rengin hangi veriye ait olduğunu bilmek için label eklenir
plt.plot(days,max_t, label="Maksimum")
plt.plot(days,min_t, label="Minimum")
plt.plot(days,avg_t, label="Ortalama")

plt.legend(shadow=True,prop={"size":10}) # label leri aktifleştirmek için
# konumunu değiştirmek için parantez içine 'upper right' 'best'
# parantez içine shadow=True de eklersem gölge verir
# yazının boyutuyla oynamak için  prop={"size":3}

plt.grid(linewidth=1)  # arkaya ızgara koyar kalınlık renk gibi özellikleri vae
```

```python
import numpy as np
```

```python
sirket=["google","amzn","msft","fb"]
kazanc=[90,136,89,27]
kar=[40,2,34,12]
```

```python
plt.bar(sirket,kazanc)
```

```python
# şirketler listesi stirng hepsi sayısal veriye çevirme len komutu ile yapılır
#ve diziye çevirdik array ile
xpos=np.arange(len(sirket))
xpos+0.2
```

```python
plt.bar(xpos+0.2,kazanc, label="ciro")  # grafiği kaydırma
plt.legend()
# x eksenindeki sayılar yerine yazılar görünsün istiyorsam
plt.xticks(xpos,sirket)
```

```python
# şirketler listesini mutlaka diziye çevirmelisin
plt.ylabel("Kazanç (milyar)")
plt.xlabel("Şirket")
plt.title("ABD teknoloji şirketleri kazançları")
plt.bar(xpos-0.2,kazanc, width=0.4, label="Ciro")
plt.bar(xpos+0.2,kar, width=0.4, label="Kar")
plt.legend()
```

```python
#plt.barh(sirket,kazanc)

# yatay eksende görmek için

plt.barh(xpos-0.2,kazanc, label="Ciro")
plt.barh(xpos+0.2,kar, label="Kar")
```

```python
kan_seker=[113,85,90,150,149,88,93,115,135,80,77,82,129]
```

```python
plt.hist(kan_seker)
```

```python
# kendimiz gruba ayırmak için bins=3 üç grup
# grafikteki veriler birbiren yapışmasın diye  rwidth kullanılır
plt.hist(kan_seker, bins=[80,100,125,150], rwidth=0.90)
```

```python
kan_seker_erkek=[113,85,90,150,149,88,93,115,135,80,77,82,129]
kan_seker_kadin=[67,98,89,120,133,150,84,69,89,79,120,112,100]

plt.hist([kan_seker_erkek,kan_seker_kadin],bins=[80,100,125,150],
         rwidth=0.95, color=["green","orange"], label=["erkek","kadin"])
plt.legend()
```

```python
#  PASTA GRAFİK
```

```python
harcamalar=[1400,600,300,410,250]
harcama_kalemleri=["Kira","Yemek","Telefon","Arac","Diğer"]
```

```python
plt.pie(harcamalar,labels=harcama_kalemleri)
plt.show()
```

```python
plt.pie(harcamalar,labels=harcama_kalemleri,
        radius=1.5,              # boyut için
        autopct="%0.2f%%",       # yüzdelik değeri için
        explode=[0,0.2,0,0,0],   # istedğin dilimi çıkartmak içn
        startangle=45            # 45 derece çevirme
        )
plt.show()
plt.savefig("dosya.png")
```
